# **1. Scatter Plot Game**

Let's create a game where random points are generated on a graph, and the player has to input the coordinates.

```python
import matplotlib.pyplot as plt
import random

def scatter_plot_game():
    score = 0
    rounds = 3  # Number of rounds to play
    xmin = -20
    xmax = 20
    ymin = -20
    ymax = 20

    for _ in range(rounds):
        xpoint = random.randint(xmin, xmax)
        ypoint = random.randint(ymin, ymax)

        fig, ax = plt.subplots()
        plt.axis([xmin, xmax, ymin, ymax])  # Set the window size
        plt.plot([xmin, xmax], [0, 0], 'b')  # x axis
        plt.plot([0, 0], [ymin, ymax], 'b')  # y axis
        plt.plot([xpoint], [ypoint], 'ro')  # plot the point
        plt.grid(True)
        plt.show()

        guess = input("Enter the coordinates of the red point (x,y): ").split(",")
        xguess = int(guess[0])
        yguess = int(guess[1])

        if xguess == xpoint and yguess == ypoint:
            score += 1

    print(f"Your score: {score}/{rounds}")

# Run the game
scatter_plot_game()
```

# **2. Algebra Practice Game**

Let's create a game with one-step and two-step algebra problems for practice.

```python
import random

def algebra_practice_game():
    score = 0
    rounds = 5  # Number of rounds to play

    for _ in range(rounds):
        problem_type = random.choice(["one-step", "two-step"])

        if problem_type == "one-step":
            a = random.randint(-10, 10)
            b = random.randint(-10, 10)
            answer = a + b
            print(f"Solve: {a} + {b} = ?")

        elif problem_type == "two-step":
            a = random.randint(-10, 10)
            b = random.randint(-10, 10)
            c = random.randint(-10, 10)
            answer = a * b + c
            print(f"Solve: {a} * {b} + {c} = ?")

        user_answer = int(input("Your answer: "))
        if user_answer == answer:
            score += 1
            print("Correct!")
        else:
            print(f"Incorrect. The correct answer was {answer}.")

    print(f"Your score: {score}/{rounds}")

# Run the game
algebra_practice_game()
```

# **3. Projectile Game**

Let's create a game where the player adjusts sliders to clear a wall.

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interactive
import random

def projectile_path(a, b, c):
    t = np.linspace(0, 10, num=1000)
    x = b * t
    y = a * t**2 + b * t + c

    # Randomize the wall
    wall_height = random.randint(5, 15)
    wall_location = random.randint(5, 10)

    fig, ax = plt.subplots()
    plt.plot(x, y, label='Projectile Path')
    plt.plot([wall_location, wall_location], [0, wall_height], 'r', label='Wall')
    plt.xlabel('Distance (m)')
    plt.ylabel('Height (m)')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title('Projectile Game')
    plt.legend()
    plt.grid(True)
    plt.show()

interactive_plot = interactive(projectile_path, a=(-10.0, 0.0, 0.1), b=(0.0, 20.0, 0.1), c=(0.0, 10.0, 0.1))
interactive_plot
```
